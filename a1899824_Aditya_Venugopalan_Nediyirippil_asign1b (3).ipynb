```python
import pandas as pd
df['date'] = pd.to_datetime(df['date'])
df.set_index('date', inplace=True)
daily_data = df.resample(rule='D').sum()
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(daily_data['Appliances'], model='additive', period=24)
fig = decomposition.plot()
plt.show()

# code reference: https://www.statsmodels.org/dev/generated/statsmodels.tsa.seasonal.seasonal_decompose.html
```

#### Conclusions

```python
from statsmodels.tsa.stattools import adfuller
adf_result = adfuller(df['Appliances'])
print('ADF Statistic: %f' % adf_result[0])
print('p-value: %f' %  adf_result[1])
if adf_result[1] < 0.05:
    print('The data is stationary (reject null hypothesis)')
else:
    print('The data is not stationary (fail to reject null hypothesis)')
#  I was able to use ADF test with reference of ChatGpt 3.5
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf 

# Reference: https://www.statsmodels.org/dev/generated/statsmodels.graphics.tsaplots.plot_acf.html
# Reference: https://www.statsmodels.org/stable/generated/statsmodels.graphics.tsaplots.plot_pacf.html

plot_acf(df2)
plot_pacf(df2)
plt.show()

#code reference https://www.machinelearningplus.com/time-series/arima-model-time-series-forecasting-python/#google_vignette
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit an ARIMA model
order = (2, 0, 1)
arima_model = ARIMA(df2, order=order)
arima_result = arima_model.fit()
print(arima_result.summary())
plt.figure(figsize=(10, 6))
plt.plot(df2, label="Original Prices")
plt.plot(arima_result.predict(), color='red', label="Predicted Values")
plt.title("ARIMA Model appliances consumption")
plt.legend()
plt.show()
# This code was able to achieve with the reference of Workshop
```

```python
# ARIMA predictions
arima_predictions = arima_result.predict(start=len(data_train), end=len(data_train)+len(data_test)-1, typ='levels')
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from math import sqrt


# Calculate metrics for ARIMA
arima_rmse = sqrt(mean_squared_error(data_test['Appliances'], arima_predictions))
arima_r2 = r2_score(data_test['Appliances'], arima_predictions)
arima_mae = mean_absolute_error(data_test['Appliances'], arima_predictions)

# Print results
print("ARIMA Metrics:")
print(f"RMSE: {arima_rmse}")
print(f"R2 Score: {arima_r2}")
print(f"MAE: {arima_mae}")
# This code was able to achieve with the reference of chatgpt 3.5 
```

### LSTM 

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense, Dropout
train_values = data_train.values
test_values = data_test.values
scaler = MinMaxScaler()
train_scaled = scaler.fit_transform(train_values)
test_scaled = scaler.transform(test_values)
time_steps = 10 
# The command prompt time_steps was taken with the help of reference of chatgpt 3.5. This helped me understand that the model will take previous 10 time steps to predict the value at the next time step.

def create_time_series(data, time_steps):
    X, y = [], []
    for i in range(len(data) - time_steps):
        X.append(data[i:(i + time_steps), 0])
        y.append(data[i + time_steps, 0])
    return np.array(X), np.array(y)


X_train, y_train = create_time_series(train_scaled, time_steps)
X_test, y_test = create_time_series(test_scaled, time_steps)

X_train = np.reshape(X_train, (X_train.shape[0], time_steps, 1))
X_test = np.reshape(X_test, (X_test.shape[0], time_steps, 1))

# Build the LSTM model
model = Sequential()
model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])))
model.add(Dropout(0.2))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))

# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Train the model
model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test), verbose=1)
# # This code was able to achieve with the reference of chatgpt 3.5 , Workshop and modules
```

```python
y_true = scaler.inverse_transform(data_test) # workshop
# Predictions for LSTM
y_pred_lstm = model.predict(X_test)

# Inverse transform predictions to original scale
y_pred_lstm = scaler.inverse_transform(y_pred_lstm)

# Calculate metrics for LSTM
lstm_rmse = sqrt(mean_squared_error(y_test, y_pred_lstm))
lstm_r2 = r2_score(y_test, y_pred_lstm)
lstm_mae = mean_absolute_error(y_test, y_pred_lstm)

# Print results
print("LSTM Metrics:")
print(f"RMSE: {lstm_rmse}")
print(f"R2 Score: {lstm_r2}")
print(f"MAE: {lstm_mae}")
# This code was able to achieve with the reference of chatgpt 3.5 
```
